|Стр. литерал | Режим |	Описание |
|-------------|-------|----------|
'r'	|Read (чтение) |Открыть файл только для чтения. Такой файл не может быть изменен.
'w' |Write (запись)	|Открыть файл для записи. Если файл уже существует, стереть его содержимое.Если файл не существует, он будет создан.
'a'	|Append (добавление) | Открыть файл для записи. Данные будут добавлены в конец файла. Если файл не существует, он будет создан.
'r+'| Read + Write | Открыть файл для чтения и записи. В этом режиме происходит частичная перезапись содержимого файла.
'x'	|Create (создание)| Создать новый файл. Если файл существует, произойдет ошибка.

```python
def read_csv():
    with open("data.csv", "r", encoding="utf-8") as f:
        
        key = f.readline().replace("\n","").split(",")
        m= [dict(zip(key,el.replace("\n", "").split(",")))for el in f.readlines()]

    return m

read_csv()
```

```python
with open("input.txt", "r", encoding="utf-8") as i_f, open("output.txt", "w", encoding="utf-8") as o_f:
    m = i_f.readlines()
    my_list = list(map(lambda i,x: f"{i}) {x}" ,list(range(1,len(m)+1)),m))
    o_f.writelines(my_list)
```

```python
with open("goats.txt", "r", encoding="utf-8") as g, open("answer.txt", "w", encoding="utf-8") as a:
    g.readline()
    my_dict = {}
    f = True
    c = 0
    for line in g:
        l = line.rstrip()
        if l == "GOATS":
            f = False
            continue
        if f:
            my_dict[l] = 0
        else:
            my_dict[l] +=1
            c +=1
    a.write("\n".join(sorted(filter(lambda x: (my_dict[x] / c) > 0.07, my_dict.keys()))))
```

```python
with open("forbidden_words.txt", "r", encoding="utf-8") as w, open(input(),'r', encoding="utf-8") as f:
    file = f.read()
    file_s = file.lower()
    for word in w.read().split():
        file_s = file_s.replace(word,"*"*len(word))
    
    for i in range(len(file)):
        if file_s[i] == "*":
            print("*",end="")
        else:
            print(file[i],end="")
```

```python
def index_of_nearest(numbers, number):
    if len(numbers) == 0:
        return -1
    c = [0,abs(numbers[0] - number)]
    for i,x in enumerate(numbers):
        a_x = abs(x-number)
        if c[1] > a_x:
            c = [i,a_x]
    return c[0]
print(index_of_nearest([12,21,2,3,42,19], 17))
```
