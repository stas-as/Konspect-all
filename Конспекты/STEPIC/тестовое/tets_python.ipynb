```python
import pandas as pd #доолжен быть еще openpyxl
```

```python
# Необходимо с файлом export.xlsx выполнить следующие действия:
df = pd.read_excel("export.xlsx")
display(df)
# - увеличить цену на 2 у всех позиций у которых в ячейке F четное число
for i in df["KOD"]:
    if df.loc[df["KOD"] == i, ["SKID"]].values[0][0]% 2 == 0: 
        df.loc[df["KOD"] == i, ["CENA"]] = df.loc[df["KOD"] == i, ["CENA"]].apply(lambda x: x+2)

```

```python
# - добавить следующие позиций в файл:

with open("тестовое задание.txt", "r", encoding="utf-8") as f:
    my_list= f.readlines()
result = filter(lambda x: x != "" and x[0] == x[-1] == "|",map(lambda x: x.replace("\n","") , my_list))

for el in result:
    df.loc[len(df.index)] = [float(x) if x.isdigit() else x for x in map(str.strip, el[1:-1].split("|"))]
```

```python
# - удалить строку 70 с кодом 421730
df = df.drop([df[df["KOD"]== 421730].index.values[0]],axis = 0)
```

```python
# приконкатенировать у всех позиций с наименованием содержашим "10" в конце цену 
for el in df["KOD"]:
    if "10" in df.loc[df["KOD"] == el,["NAME"]].values[0][0]:
        df.loc[df["KOD"] == el,["NAME"]] = f'{df.loc[df["KOD"] == el,["NAME"]].values[0][0]} {str(df.loc[df["KOD"]== el,["CENA"]].values[0][0])}'

display(df)
```

```python
# отсортировать файл по цене (ячейка C)

df = df.sort_values("CENA")
display(df)
```
