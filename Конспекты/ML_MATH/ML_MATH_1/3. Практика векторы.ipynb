**?**

**Задача (условие № 1)**

Агентство «Рай в Шалаше» недавно сдало в аренду четыре квартиры в одном доме.

* Общая площадь: 33, 42, 33 и 65 кв. м.
* Жилая площадь: 24, 32, 20 и 48 кв. м.
* Арендная плата: 39, 50, 38 и 77 тысяч рублей соответственно.

В том же доме появилась новая квартира общей площадью 50 кв. м, из которых жилая — 40 кв. м.

На основе имеющихся данных выберите оптимальную арендную плату для новой квартиры.

Запишем данные в векторном виде.

Вектор общих площадей: $${\overrightarrow{s}}_{\textrm{о}\textrm{б}\textrm{щ}}=\left( \begin{array}{c} 33 \\ \mathrm{42} \\ 33 \\ 65 \end{array} \right)\ [\textrm{к}\textrm{в}.\ \textrm{м}]$$

Вектор жилых площадей: $${\overrightarrow{s}}_{\textrm{ж}\textrm{и}\textrm{л}}=\left( \begin{array}{c} 24 \\ \mathrm{32} \\ 20 \\ 48 \end{array} \right)\ [\textrm{к}\textrm{в}.\ \textrm{м}]$$

Стоимость аренды: $$\overrightarrow{p}=\left( \begin{array}{c} 39 \\ \mathrm{50} \\ 38 \\ 77 \end{array} \right)\ [\textrm{т}\textrm{ы}\textrm{с}.\ \textrm{р}\textrm{у}\textrm{б}]$$

Нам нужно порекомендовать оптимальную арендную плату для новой квартиры, исходя из опыта аренды предыдущих четырёх квартир.

Ключом к решению является линейная зависимость векторов: можно заметить, что вектор $\overrightarrow{p}$ является линейной комбинацией векторов ${\overrightarrow{s}}_{\textrm{о}\textrm{б}\textrm{щ}}$ и ${\overrightarrow{s}}_{\textrm{ж}\textrm{и}\textrm{л}}$ с коэффициентами ${\omega }_1=1$ и ${\omega }_2=\frac{1}{4}$:

$$\overrightarrow{p}={\omega }_1{\overrightarrow{s}}_{\textrm{о}\textrm{б}\textrm{щ}}+{\omega }_2{\overrightarrow{s}}_{\textrm{ж}\textrm{и}\textrm{л}}=\left( \begin{array}{c} 33 \\ \mathrm{42} \\ 33 \\ 65 \end{array} \right)+\frac{1}{4}\left( \begin{array}{c} 24 \\ \mathrm{32} \\ 20 \\ 48 \end{array} \right)=\left( \begin{array}{c} 33 \\ \mathrm{42} \\ 33 \\ 65 \end{array} \right)+\left( \begin{array}{c} 6 \\ \mathrm{8} \\ 5 \\ 14 \end{array} \right)=\left( \begin{array}{c} 39 \\ \mathrm{50} \\ 38 \\ 77 \end{array} \right)[\textrm{т}\textrm{ы}\textrm{с}.\ \textrm{р}\textrm{у}\textrm{б}]\$$

Так как новая квартира расположена в том же доме, что и первые четыре, она находится в тех же условиях, то оптимальную стоимость найти несложно.

Если вектор $\overrightarrow{p}$ является линейной комбинацией векторов ${\overrightarrow{s}}_{\textrm{о}\textrm{б}\textrm{щ}}$ и ${\overrightarrow{s}}_{\textrm{ж}\textrm{и}\textrm{л}}$, то для того, чтобы определить оптимальную цену нового жилья, нам нужно просто сложить общую и жилую площади с коэффициентами ${\omega }_1=1$ и ${\omega }_2=\frac{1}{4}$. В таком случае цена аренды новой квартиры будет полностью соответствовать остальным квартирам.

Составим вектор из коэффициентов $w$ линейной комбинации $\overrightarrow{\omega }=\left( \begin{array}{c} {\omega }_1 \\ {\omega }_2 \end{array} \right)=\left( \begin{array}{c} 1 \\ \frac{1}{4} \end{array} \right)$ и  скалярно умножим его на вектор новой квартиры $\overrightarrow{x}=\left( \begin{array}{c} 50 \\ 40 \end{array} \right)$:

$$p=\left(\overrightarrow{\omega },\overrightarrow{x}\right)=\ 50+\frac{1}{4}40=60\ [\textrm{т}\textrm{ы}\textrm{с}.\ \textrm{р}\textrm{у}\textrm{б}]$$

Не кажется ли вам, что случай, когда вектор $\overrightarrow{p}$ так просто выражается через векторы ${\overrightarrow{s}}_{\textrm{о}\textrm{б}\textrm{щ}}$ и ${\overrightarrow{s}}_{\textrm{ж}\textrm{и}\textrm{л}}$ слишком прост для реальности? Давайте попробуем решить более приближённую к реальным данными задачу.

**?**

**Задача (условие № 2)**

Пусть во второй квартире обнаружились бытовые проблемы, и её пришлось сдать дешевле (за 48 460 рублей), а в первой и третьей квартирах был застеклённый балкон, и их удалось сдать дороже (за 40 000 и за 39 000 рублей соответственно).

На основе имеющихся данных выберите оптимальную арендную плату для новой квартиры.

Вектор общих площадей: 

$${\overrightarrow{s}}_{\textrm{о}\textrm{б}\textrm{щ}}=\left( \begin{array}{c} 33 \\ \mathrm{42} \\ 33 \\ 65 \end{array} \right)\ [\textrm{к}\textrm{в}.\ \textrm{м}]$$

Вектор жилых площадей: 

$${\overrightarrow{s}}_{\textrm{ж}\textrm{и}\textrm{л}}=\left( \begin{array}{c} 24 \\ \mathrm{32} \\ 20 \\ 48 \end{array} \right)\ [\textrm{к}\textrm{в}.\ \textrm{м}]$$

Стоимость аренды: 

$${\overrightarrow{p}}_{\textrm{р}\textrm{е}\textrm{а}\textrm{л}}=\left( \begin{array}{c} 40 \\ \mathrm{48.46} \\ 39 \\ 77 \end{array} \right)\ [\textrm{т}\textrm{ы}\textrm{с}.\ \textrm{р}\textrm{у}\textrm{б}]$$

Теперь вектор арендной платы не выражается линейно через векторы площадей.

Изобразим данные геометрически. Так как векторы четырёхмерные, мы не сможем получить корректное геометрическое решение. Однако можно получить условное графическое представление, которого нам будет достаточно для решения задачи.

Векторы общей и жилой площади лежат в некоторой плоскости, обозначенной синим цветом. Вектор $\overrightarrow{p}$ из первого условия задачи линейно выражается через них, а значит, лежит в той же плоскости. 

Вектор ${\overrightarrow{p}}_{\textrm{р}\textrm{е}\textrm{а}\textrm{л}}$ выразить через векторы площадей уже нельзя, а значит он не лежит в синей плоскости.



Красный вектор $\overrightarrow{n}$ как разница между ${\overrightarrow{p}}_{\textrm{р}\textrm{е}\textrm{а}\textrm{л}}$ и $\overrightarrow{p}$ символизирует влияние других, внешних, факторов на стоимость аренды. Вычислим его координаты и найдём скалярное произведение с вектором $\overrightarrow{p}$:



Скалярное произведение равно нулю. Это значит, что угол между векторами $\overrightarrow{p}$ и $\overrightarrow{n}$ равен 90 градусов, то есть $\overrightarrow{p}\bot \overrightarrow{n}$. Говорят: вектор является ортогональной проекцией вектора $\overrightarrow{p}$ на плоскость.

Но что это значит для нас?

Это значит, что среди всех возможных линейных комбинаций векторов общей и жилой площади наше изначальное ${\overrightarrow{p}}$ — наилучший прогноз арендной платы. Подумайте, почему.

⭐ А теперь кульминация задачи, которую мы только что решили ↓

На самом деле, решая данную задачу, мы с вами только что построили простейшую **модель линейной регрессии в векторном виде**. Давайте посмотрим на задачу под другим углом.

У нас было:

* два вектора ${\overrightarrow{s}}_{\textrm{о}\textrm{б}\textrm{щ}}$ и ${\overrightarrow{s}}_{\textrm{ж}\textrm{и}\textrm{л}}$ — признаки (факторы), которые описывают объекты в данных;
* вектор ${\overrightarrow{p}}_{\textrm{р}\textrm{е}\textrm{а}\textrm{л}}$ — целевой признак, который мы пытаемся предсказать.

Мы построили такую линейную комбинацию $\overrightarrow{p}$ из векторов ${\overrightarrow{s}}_{\textrm{о}\textrm{б}\textrm{щ}}$ и ${\overrightarrow{s}}_{\textrm{ж}\textrm{и}\textrm{л}}$, при которой ошибка между вектором и линейной комбинацией $\overrightarrow{e}{=\overrightarrow{p}}_{\textrm{р}\textrm{е}\textrm{а}\textrm{л}}-\overrightarrow{p}$ была бы минимальной.

Коэффициенты линейной комбинации ${\omega }_1$ и ${\omega }_2$ —это наши любимые коэффициенты линейной регрессии (правда, в классической линейной регрессии есть ещё коэффициент добавочного смещения ${\omega }_0$, но эту задачу мы решили без его участия).

**?**
У вас, должно быть, возник вопрос: как ищутся коэффициенты линейной комбинации ${\omega }_1$ и ${\omega }_2$? Не искать же их полным перебором.

На этот вопрос в своё время ответил Гаусс и создал знакомый нам **метод наименьших квадратов (МНК)**, который мы уже многократно использовали в модулях по машинному обучению. Но, чтобы познакомиться с МНК поближе с математической точки зрения, нам с вами необходимо изучить тему **матриц** и **матричных преобразований.**

**Задача (условие № 3)**

Пусть появилась дополнительная информация — нежилая площадь.

Позволит ли она улучшить прогноз по аренде новой квартиры?

Вектор общих площадей: 

$${\overrightarrow{s}}_{\textrm{о}\textrm{б}\textrm{щ}}=\left( \begin{array}{c} 33 \\ \mathrm{42} \\ 33 \\ 65 \end{array} \right)\ [\textrm{к}\textrm{в}.\ \textrm{м}]$$

Вектор жилых площадей: 

$${\overrightarrow{s}}_{\textrm{ж}\textrm{и}\textrm{л}}=\left( \begin{array}{c} 24 \\ \mathrm{32} \\ 20 \\ 48 \end{array} \right)\ [\textrm{к}\textrm{в}.\ \textrm{м}]$$

Вектор нежилых площадей: 

$${\overrightarrow{s}}_{\textrm{н}\textrm{ж}}=\left( \begin{array}{c} 9 \\ \mathrm{10} \\ 13 \\ 17 \end{array} \right)\ [\textrm{к}\textrm{в}.\ \textrm{м}]$$

Стоимость аренды: 

$${\overrightarrow{p}}_{\textrm{р}\textrm{е}\textrm{а}\textrm{л}}=\left( \begin{array}{c} 40 \\ \mathrm{48.46} \\ 39 \\ 77 \end{array} \right)\ [\textrm{т}\textrm{ы}\textrm{с}.\ \textrm{р}\textrm{у}\textrm{б}]$$

Нежилая площадь является разностью общей и жилой, а значит, лежит в той же плоскости. То есть вектор ${\overrightarrow{s}}_{\textrm{н}\textrm{ж}}$ линейно выражается через векторы ${\overrightarrow{s}}_{\textrm{о}\textrm{б}\textrm{щ}}$ и ${\overrightarrow{s}}_{\textrm{ж}\textrm{и}\textrm{л}}$.



А это значит, что проекция ${\overrightarrow{p}}_{\mathrm{\textrm{р}\textrm{е}\textrm{а}\textrm{л}}}$ на эту плоскость не поменяется, то есть и наилучший прогноз останется без изменений.



Линейные комбинации факторов не добавляют точности прогнозу и, как мы увидим позже, сильно портят вычисления при построении линейных регрессий. Именно поэтому вопрос о линейной зависимости векторов имеет крайне важное значение при построении регрессий.

>**Длина вектора** (в физическом смысле) — это то же, что и норма, применимо к векторам в реальном пространстве.

Нормирование производится делением вектора на его норму.
